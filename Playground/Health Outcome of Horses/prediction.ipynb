```python
import pandas as pd
```

```python
train_df = pd.read_csv('simplified_trained_df.csv')
train_df.head()
```

```python
test_df = pd.read_csv('test.csv')
test_df.head()
```

```python
test_df = test_df.drop(columns=['id', 'hospital_number'])
```

```python
test_df.isna().sum()
```

```python
train_df.isna().sum()
```

```python
numerical_columns = []
categorical_columns = []
for cols in test_df:
    if test_df[cols].dtype == 'object':
        categorical_columns.append(cols);
    else:
        numerical_columns.append(cols);
```

```python
categorical_columns
```

```python
numerical_columns
```

```python
for col in test_df.columns:
    if col in categorical_columns:
        mode_value = test_df[col].mode()[0]  
        test_df[col].fillna(mode_value, inplace=True)  
    else:
        median_value = test_df[col].median()  
        test_df[col].fillna(median_value, inplace=True) 
```

```python
test_df.isna().sum()
```

```python
X = test_df
```

```python
X.shape
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
categorical_features = ['surgery',
 'age',
 'temp_of_extremities',
 'peripheral_pulse',
 'mucous_membrane',
 'capillary_refill_time',
 'pain',
 'peristalsis',
 'abdominal_distention',
 'nasogastric_tube',
 'nasogastric_reflux',
 'rectal_exam_feces',
 'abdomen',
 'abdomo_appearance',
 'surgical_lesion',
 'cp_data']
numerical_features = ['rectal_temp',
 'pulse',
 'respiratory_rate',
 'nasogastric_reflux_ph',
 'packed_cell_volume',
 'total_protein',
 'abdomo_protein',
 'lesion_1',
 'lesion_2',
 'lesion_3']
ct = ColumnTransformer([
    ('OneHotEncoder', OneHotEncoder(), categorical_features),
    ('StandardScaling', StandardScaler(), numerical_features)
])
X = ct.fit_transform(X)
X.shape
```
